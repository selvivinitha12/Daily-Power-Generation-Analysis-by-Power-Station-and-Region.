```python
# Check missing values
print(df.isnull().sum())

# Fill missing values with mode
df["sector"] = df["sector"].fillna(df["sector"].mode()[0])
df["station_type"] = df["station_type"].fillna(df["station_type"].mode()[0])

```

```python
df = df.rename(columns={
    "state_name": "state",
    "state_code": "state_code",
    "power_station": "station_name",
    "power_station_unit": "station_unit",
    "todays_gen_prgm": "planned_generation",
    "todays_gen_act": "actual_generation"
})

print(df.columns)
```

```python
Q1 = df['monitored_capacity'].quantile(0.25)
Q3 = df['monitored_capacity'].quantile(0.75)
IQR = Q3 - Q1

```

```python
df = df[(df['monitored_capacity'] < (Q1 - 1.5 * IQR)) |
             (df['monitored_capacity'] > (Q3 + 1.5 * IQR))]
df
```

```python
df["capacity_utilization"] = (
    df["actual_generation"] / df["monitored_capacity"]
) * 100

df[["monitored_capacity", "actual_generation", "capacity_utilization"]].head()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x='region', data=df)
plt.title("count of records by region")
plt.xticks(rotation=45)
plt.show()
```



*   Western Region has the highest number of records (approximately 245), indicating that it contributes the largest share of data in the dataset.
*   Northern Region ranks second with approximately 205 records, showing strong representation after the Western region.


*   Eastern Region has the lowest number of records (approximately 40), making it the least represented region in the dataset.
*   The dataset is imbalanced across regions, with the majority of records concentrated in the Western and Northern regions, while the Southern and Eastern regions have comparatively fewer observations. This imbalance should be considered during further analysis and model development.












```python
plt.figure(figsize=(8,5))
plt.hist(df['actual_generation'], bins=20)
plt.title("Distribution of Actual Power Generation")
plt.xlabel("Actual Generation")
plt.ylabel("Frequency")
plt.show()
```



*   Most power generation values are very low, with the highest frequency concentrated between 0 and 2 units, indicating that the dataset contains many low-generation observations.
*   The distribution is positively (right) skewed, as a small number of observations extend to higher generation values (up to around 37 units)


*  Power generation values are widely spread, with noticeable peaks around 10–13 units and 20–23 units, suggesting variability in generation levels across observations.
*   A few high-value observations (above 30 units) appear rarely, indicating the presence of potential outliers or periods of exceptionally high power generation.





```python
sns.countplot(data=df, x="station_type", hue="sector")
plt.title("Station Type Vs Sector")
plt.xlabel("Station Type")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```



*  Hydro stations are predominantly owned by the State sector, making them the most common station type across all sectors.
*   Thermal (GT) stations are more concentrated in the Private sector, while the State and Central sectors have comparatively fewer Thermal stations.


*   Nuclear stations are found only in the Central sector, indicating that nuclear power generation is exclusively managed by the Central government in this dataset.






```python
sns.heatmap(df.corr(numeric_only=True),annot=True)
plt.show()
```



*   Actual Generation and Capacity Utilization have a very strong positive correlation (0.96), indicating that higher power generation is associated with higher utilization of plant capacity.
*   Planned Generation shows a moderate positive correlation with Actual Generation (0.46) and Capacity Utilization (0.44), suggesting that higher planned output generally results in higher actual generation and utilization.

*  Monitored Capacity has a weak correlation with the other variables, while State Code has a moderate negative correlation with Actual Generation (-0.43), indicating that generation levels vary across different states in the dataset.





```python
plt.figure(figsize=(10,5))

sns.lineplot(data=df,
             x='date',
             y='actual_generation',
             hue='region',
             marker='o')

plt.title("Daily Actual Generation Trend by Region")
plt.xlabel("Date")
plt.ylabel("Actual Generation (MW)")
plt.xticks(rotation=45)
plt.show()
```





*  The Eastern region consistently records the highest actual power generation, maintaining around 21–22 MW on most days, while the Western region has the lowest generation.
*   All regions experience a sharp drop to nearly zero around September 22, indicating a possible outage, maintenance activity, or missing data across the dataset.


*  The Northern, Southern, and Western regions show moderate day-to-day fluctuations, whereas the Eastern region remains relatively stable, indicating more consistent power generation over the observed period.













The violin plot shows both the spread and density of utilization values.
A narrow shape indicates consistent plant performance.
A wide shape suggests fluctuations in operational efficiency.
Management can investigate regions with inconsistent utilization.

```python
plt.figure(figsize=(12,5))

sns.stripplot(data=df,
              x='station_name',
              y='actual_generation',
              jitter=True)

plt.title("Actual Generation by Power Station")
plt.xlabel("Power Station")
plt.ylabel("Actual Generation (MW)")
plt.xticks(rotation=90)
plt.show()
```

Each point represents a day's electricity generation.
Stations with tightly grouped points are more consistent.
Large spreads indicate unstable generation performance.
Consistent stations are generally easier to plan and manage.

```python
sns.pairplot(df[['monitored_capacity',
                 'actual_generation',
                 'capacity_utilization']],
             diag_kind='hist')

plt.show()
```



*  Actual Generation and Capacity Utilization show a strong positive linear relationship, indicating that higher power generation leads to higher capacity utilization.
*  Monitored Capacity has no strong linear relationship with Actual Generation or Capacity Utilization, as the data points are widely scattered across different capacity levels.

*   The diagonal histograms show that Monitored Capacity is concentrated around a few capacity levels, while Actual Generation and Capacity Utilization have a wider distribution.
*   A few extreme values are visible in Actual Generation and Capacity Utilization, suggesting the presence of potential outliers or periods of unusually high plant performance.
